```python
from faker import Faker
import pandas as pd
import numpy as np
from datetime import datetime
import random
fk=Faker()

#for actives data
advertising_id=[]
city =[]
location_category = []
location_granularities = []
location_source= []
state =[]
timestamp = []
user_id = []
user_latitude = []
user_longitude = []
date = []

for i in range(100):
    advertising_id.append(str(fk.unique.random_int()))
    city.append(fk.city())
    location_category.append(str(fk.local_latlng()[3]))
    location_granularities.append(str(fk.local_latlng()[4]))
    location_source.append(str(fk.local_latlng()[2:])[1:-1])
    state.append(fk.state())
    timestamp.append(str(fk.time()))
    user_id.append(fk.bothify('???-???-####'))
    user_latitude.append(float(fk.latitude()))
    user_longitude.append(float(fk.longitude()))
    date.append(datetime.strptime(fk.date(),'%Y-%m-%d'))

    
Actives=pd.DataFrame(zip(advertising_id,city,location_category,location_granularities,location_source,state,timestamp,user_id,user_latitude,user_longitude,date),columns=['advertising_id','city','location_category','location_granularities','location_source','state','timestamp','user_id','user_latitude','user_longitude','date'])

Actives['month']=Actives['date'].dt.strftime('%B')


#for viewsership data
advertising_id2=[]
channel_genre=[]
channel_name=[]
city2=[]
device=[]
device_type=[]
duration=[]
grid_id=[]
language=[]
location_category2=[]
location_granularities2=[]
location_source2=[]
record_timestamp2=[]
show_genre=[]
show_name=[]
state2=[]
user_lat=[]
user_long=[]
month=[]
date2=[]

for i in range(99):
    advertising_id2.append(str(random.choice(advertising_id))) #Restricted to the already created ids
    channel_genre.append(random.choice(['News','Cartoon','Movies','Music','Reality']))
    channel_name.append(fk.safe_color_name())
    city2.append(random.choice(city))
    device.append(fk.ssn())
    device_type.append(random.choice(['mobile','TV','others']))
    duration.append(random.randint(1,400))
    grid_id.append(fk.bothify('?#-##'))
    language.append(random.choice(['Brazilian Portuguese','Canadian French','English','French Creole','Quechua','Spanish']))
    location_category2.append(random.choice(location_category))
    location_granularities2.append(random.choice(location_granularities))
    location_source2.append(random.choice(location_source))
    record_timestamp2.append(random.choice(timestamp))
    show_genre.append(random.choice(['Action','Adventure','Horror','Comedy','Romance']))
    show_name.append(fk.job())
    state2.append(random.choice(state))
    user_lat.append(random.choice(user_latitude))
    user_long.append(random.choice(user_longitude))
    date2.append(random.choice(date))
    
    
Viewvership=pd.DataFrame(zip(advertising_id2,channel_genre,channel_name,city2,device,device_type,duration,grid_id,language,location_category2,location_granularities2,location_source2,record_timestamp2,show_genre,show_name,state2,user_lat,user_longitude,date2),columns=['advertising_id','channel_genre','channel_name','city','device','device_type','duration','grid_id','language','location_category','location_granularities','location_source','record_timestamp','show_genre','show_name','state','user_lat','user_long','date'])    
Viewvership['month']=Viewvership['date'].dt.strftime('%B')

Actives.head(3)
# Viewvership.info()
# Actives.info()
```
